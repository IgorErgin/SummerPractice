```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import seaborn as sns
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader

# Устанавливаем устройство (GPU, если доступно, иначе CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Загружаем набор данных MNIST
def get_mnist_loaders(batch_size):
    # Определяем преобразования для изображений: конвертация в тензор и нормализация
    transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
    # Загружаем обучающий набор данных MNIST
    train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
    # Загружаем тестовый набор данных MNIST
    test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)
    # Создаем загрузчики данных для обучающего и тестового наборов
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    return train_loader, test_loader

# Получаем загрузчики данных MNIST
train_loader, test_loader = get_mnist_loaders(batch_size=64)

# Определяем простую полносвязную модель
class FullyConnectedModel(nn.Module):
    def __init__(self, input_size, num_classes, layers):
        super(FullyConnectedModel, self).__init__()
        self.layers = nn.ModuleList() # Список для хранения слоев модели
        prev_size = input_size
        # Проходим по всем слоям, определенным в конфигурации
        for layer in layers:
            if layer["type"] == "linear":
                # Добавляем линейный слой
                self.layers.append(nn.Linear(prev_size, layer["size"]))
                prev_size = layer["size"]
            elif layer["type"] == "relu":
                # Добавляем функцию активации ReLU
                self.layers.append(nn.ReLU())
            elif layer["type"] == "dropout":
                # Добавляем слой Dropout
                self.layers.append(nn.Dropout(p=layer["rate"]))
            elif layer["type"] == "batch_norm":
                # Добавляем слой Batch Normalization
                self.layers.append(nn.BatchNorm1d(prev_size))
        # Добавляем выходной линейный слой
        self.layers.append(nn.Linear(prev_size, num_classes))

    def forward(self, x):
        x = x.view(x.size(0), -1)  # Разворачиваем входной тензор в одномерный вектор
        for layer in self.layers:
            x = layer(x)
        return x

# Функция обучения модели
def train_model(model, train_loader, test_loader, epochs, device, optimizer):
    criterion = nn.CrossEntropyLoss() # Определяем функцию потерь (кросс-энтропия)
    history = {'train_loss': [], 'train_acc': [], 'test_acc': []} # Словарь для хранения истории обучения
    for epoch in range(epochs):
        model.train() # Переводим модель в режим обучения
        running_loss, correct, total = 0.0, 0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device) # Перемещаем данные на устройство
            optimizer.zero_grad() # Обнуляем градиенты
            outputs = model(inputs) # Прямой проход
            loss = criterion(outputs, labels) # Вычисляем потери
            loss.backward() # Обратный проход (вычисление градиентов)
            optimizer.step() # Обновление весов модели
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
        train_loss = running_loss / len(train_loader)
        train_acc = 100. * correct / total
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)

        model.eval() # Переводим модель в режим оценки
        correct, total = 0, 0
        with torch.no_grad(): # Отключаем вычисление градиентов
            for inputs, labels in test_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = outputs.max(1)
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()
        test_acc = 100. * correct / total
        history['test_acc'].append(test_acc)
        print(f'Эпоха {epoch+1}/{epochs}, Потери: {train_loss:.4f}, Точность на обучении: {train_acc:.2f}%, Точность на тесте: {test_acc:.2f}%')
    return history

# Функция для построения графиков истории обучения
def plot_training_history(history, title):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Потери на обучении')
    plt.title(f'{title} - Потери')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Точность на обучении')
    plt.plot(history['test_acc'], label='Точность на тесте')
    plt.title(f'{title} - Точность')
    plt.legend()
    plt.show()

# Базовая архитектура
base_layers = [
    {"type": "linear", "size": 256},
    {"type": "relu"},
    {"type": "linear", "size": 256},
    {"type": "relu"},
    {"type": "linear", "size": 256},
    {"type": "relu"}
]

# Вспомогательные функции для добавления регуляризации
def add_dropout(layers, rate):
    new_layers = []
    for layer in layers:
        new_layers.append(layer)
        if layer["type"] == "relu":
            new_layers.append({"type": "dropout", "rate": rate}) # Добавляем Dropout после ReLU
    return new_layers

def add_batchnorm(layers):
    new_layers = []
    for layer in layers:
        if layer["type"] == "linear":
            new_layers.append(layer)
            new_layers.append({"type": "batch_norm"}) # Добавляем BatchNorm после линейного слоя
        else:
            new_layers.append(layer)
    return new_layers

# 3.1 Сравнение методов регуляризации
print("=== Эксперимент 3.1: Сравнение методов регуляризации ===")

models = {
    "Без регуляризации": FullyConnectedModel(784, 10, base_layers).to(device),
    "Dropout 0.1": FullyConnectedModel(784, 10, add_dropout(base_layers, 0.1)).to(device),
    "Dropout 0.3": FullyConnectedModel(784, 10, add_dropout(base_layers, 0.3)).to(device),
    "Dropout 0.5": FullyConnectedModel(784, 10, add_dropout(base_layers, 0.5)).to(device),
    "BatchNorm": FullyConnectedModel(784, 10, add_batchnorm(base_layers)).to(device),
    "Dropout 0.3 + BatchNorm": FullyConnectedModel(784, 10, add_dropout(add_batchnorm(base_layers), 0.3)).to(device),
    "L2 Regularization": FullyConnectedModel(784, 10, base_layers).to(device)
}

histories = {}
for name, model in models.items():
    print(f"\nОбучение {name}...")
    # Для L2-регуляризации используем параметр weight_decay в оптимизаторе
    optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=0.001 if name == "L2 Regularization" else 0.0)
    history = train_model(model, train_loader, test_loader, epochs=10, device=device, optimizer=optimizer)
    histories[name] = history
    plot_training_history(history, name)

# Визуализация распределения весов
plt.figure(figsize=(12, 8))
for name, model in models.items():
    # Собираем все веса модели в один массив
    weights = torch.cat([p.view(-1) for p in model.parameters()]).cpu().detach().numpy()
    sns.histplot(weights, label=name, bins=50, alpha=0.5) # Строим гистограмму распределения весов
plt.legend()
plt.title('Распределения весов - 3.1')
plt.show()

# 3.2 Адаптивная регуляризация
print("\n=== Эксперимент 3.2: Адаптивная регуляризация ===")

# Адаптивный Dropout
class ScheduledDropout(nn.Module):
    def __init__(self, initial_rate, final_rate, total_epochs):
        super().__init__()
        self.initial_rate = initial_rate
        self.final_rate = final_rate
        self.total_epochs = total_epochs
        self.current_epoch = 0

    def forward(self, x):
        # Вычисляем текущую скорость Dropout линейно от initial_rate до final_rate
        rate = self.initial_rate + (self.final_rate - self.initial_rate) * (self.current_epoch / self.total_epochs)
        return nn.functional.dropout(x, p=rate, training=self.training)

    def step(self):
        # Увеличиваем номер текущей эпохи
        self.current_epoch += 1

# Пользовательская модель с адаптивными методами
class AdaptiveModel(nn.Module):
    def __init__(self, input_size, num_classes, use_adaptive_dropout=False, momentum=0.1):
        super(AdaptiveModel, self).__init__()
        self.layers = nn.ModuleList()
        prev_size = input_size
        for i in range(3): # Создаем 3 скрытых слоя
            self.layers.append(nn.Linear(prev_size, 256))
            self.layers.append(nn.BatchNorm1d(256, momentum=momentum)) # BatchNorm с настраиваемым momentum
            if use_adaptive_dropout:
                self.layers.append(ScheduledDropout(0.5, 0.1, 10)) # Адаптивный Dropout
            else:
                self.layers.append(nn.Dropout(0.3)) # Обычный Dropout
            self.layers.append(nn.ReLU())
            prev_size = 256
        self.layers.append(nn.Linear(prev_size, num_classes))

    def forward(self, x):
        x = x.view(x.size(0), -1)
        for layer in self.layers:
            x = layer(x)
        return x

    def step_dropout(self):
        # Вызываем step() для всех слоев ScheduledDropout в модели
        for layer in self.layers:
            if isinstance(layer, ScheduledDropout):
                layer.step()

adaptive_models = {
    "Адаптивный Dropout": AdaptiveModel(784, 10, use_adaptive_dropout=True).to(device),
    "BatchNorm Momentum 0.1": AdaptiveModel(784, 10, momentum=0.1).to(device),
    "BatchNorm Momentum 0.9": AdaptiveModel(784, 10, momentum=0.9).to(device),
    "Комбинированный": AdaptiveModel(784, 10, use_adaptive_dropout=True, momentum=0.5).to(device)
}

adaptive_histories = {}
for name, model in adaptive_models.items():
    print(f"\nОбучение {name}...")
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    history = train_model(model, train_loader, test_loader, epochs=10, device=device, optimizer=optimizer)
    if "Адаптивный Dropout" in name: # Если используется адаптивный Dropout, вызываем step_dropout
        model.step_dropout()
    adaptive_histories[name] = history
    plot_training_history(history, name)

# Анализ весов слоев для комбинированной модели
combined_model = adaptive_models["Комбинированный"]
# Собираем веса только из линейных слоев
layer_weights = [layer.weight.view(-1).cpu().detach().numpy() for layer in combined_model.layers if isinstance(layer, nn.Linear)]
plt.figure(figsize=(12, 8))
for i, weights in enumerate(layer_weights):
    sns.histplot(weights, label=f'Слой {i+1}', bins=50, alpha=0.5) # Строим гистограмму распределения весов для каждого слоя
plt.legend()
plt.title('Распределения весов по слоям - Комбинированная модель (3.2)')
plt.show()
```
