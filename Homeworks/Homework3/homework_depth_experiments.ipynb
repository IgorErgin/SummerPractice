```python
import torch
import time
from datasets import get_mnist_loaders
from models import FullyConnectedModel
from trainer import train_model
from utils import plot_training_history

# Определяем устройство для обучения (GPU, если доступно, иначе CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Загружаем набор данных MNIST
train_loader, test_loader = get_mnist_loaders(batch_size=64)

# Определяем модели с разной глубиной
models = {
    "1_layer": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[]
    ).to(device),
    "2_layers": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[
            {"type": "linear", "size": 256},
            {"type": "relu"}
        ]
    ).to(device),
    "3_layers": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"}
        ]
    ).to(device),
    "5_layers": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"}
        ]
    ).to(device),
    "7_layers": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"},
            {"type": "linear", "size": 256},
            {"type": "relu"}
        ]
    ).to(device)
}

# Определяем модели с регуляризацией (Dropout и BatchNorm)
models_reg = {
    "2_layers_reg": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2}
        ]
    ).to(device),
    "3_layers_reg": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2}
        ]
    ).to(device),
    "5_layers_reg": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2}
        ]
    ).to(device),
    "7_layers_reg": FullyConnectedModel(
        input_size=784,
        num_classes=10,
        layers=[
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2},
            {"type": "linear", "size": 256},
            {"type": "batch_norm"},
            {"type": "relu"},
            {"type": "dropout", "rate": 0.2}
        ]
    ).to(device)
}

# Функция для обучения и оценки моделей
def train_and_evaluate(models, train_loader, test_loader, epochs=10):
    histories = {}
    training_times = {}
    
    for name, model in models.items():
        print(f"\nОбучение {name}...")
        start_time = time.time()
        history = train_model(model, train_loader, test_loader, epochs=epochs, device=device)
        end_time = time.time()
        
        training_time = end_time - start_time
        histories[name] = history
        training_times[name] = training_time
        
        print(f"{name} - Время обучения: {training_time:.2f} секунд")
        print(f"{name} - Итоговая точность на тестовом наборе: {history['test_accs'][-1]:.4f}")
        plot_training_history(history)
    
    return histories, training_times

# 1.1 Сравнение моделей разной глубины
print("=== Эксперимент 1.1: Сравнение моделей разной глубины ===")
histories_base, times_base = train_and_evaluate(models, train_loader, test_loader)

# 1.2 Анализ переобучения с регуляризацией
print("\n=== Эксперимент 1.2: Анализ переобучения с регуляризацией ===")
histories_reg, times_reg = train_and_evaluate(models_reg, train_loader, test_loader)

# Сводный анализ
print("\n=== Сводный анализ ===")
print("\nБазовые модели:")
for name in models.keys():
    print(f"{name}:")
    print(f"  Итоговая точность на обучающем наборе: {histories_base[name]['train_accs'][-1]:.4f}")
    print(f"  Итоговая точность на тестовом наборе: {histories_base[name]['test_accs'][-1]:.4f}")
    print(f"  Время обучения: {times_base[name]:.2f} секунд")

print("\nМодели с регуляризацией:")
for name in models_reg.keys():
    print(f"{name}:")
    print(f"  Итоговая точность на обучающем наборе: {histories_reg[name]['train_accs'][-1]:.4f}")
    print(f"  Итоговая точность на тестовом наборе: {histories_reg[name]['test_accs'][-1]:.4f}")
    print(f"  Время обучения: {times_reg[name]:.2f} секунд")

# Анализ оптимальной глубины
print("\nАнализ оптимальной глубины:")
best_test_acc = 0
best_model = ""
for name in models.keys():
    test_acc = histories_base[name]['test_accs'][-1]
    if test_acc > best_test_acc:
        best_test_acc = test_acc
        best_model = name
print(f"Лучшая базовая модель: {best_model} с точностью на тестовом наборе: {best_test_acc:.4f}")

best_test_acc_reg = 0
best_model_reg = ""
for name in models_reg.keys():
    test_acc = histories_reg[name]['test_accs'][-1]
    if test_acc > best_test_acc_reg:
        best_test_acc_reg = test_acc
        best_model_reg = name
print(f"Лучшая модель с регуляризацией: {best_model_reg} с точностью на тестовом наборе: {best_test_acc_reg:.4f}")
```
