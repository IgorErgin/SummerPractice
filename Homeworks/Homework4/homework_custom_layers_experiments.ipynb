```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Function
from datasets import get_cifar_loaders
from trainer import train_model
from utils import plot_training_history, count_parameters
import time
import numpy as np
import matplotlib.pyplot as plt
import torch.optim as optim

# 3.1 Пользовательские слои
# Пользовательская функция свертки с нестандартным поведением
class CustomConvFunction(Function):
    @staticmethod
    def forward(ctx, input, weight, bias, stride=1, padding=1):
        # Сохраняем тензоры для использования в обратном проходе
        ctx.save_for_backward(input, weight, bias)
        ctx.stride = stride
        ctx.padding = padding
        # Выполняем стандартную 2D свертку
        output = F.conv2d(input, weight, bias, stride=stride, padding=padding)
        # Применяем пользовательское масштабирование: масштабируем выход по среднему значению канала
        channel_means = output.mean(dim=(2, 3), keepdim=True)
        output = output / (channel_means + 1e-5) # Добавляем малое значение для стабильности
        return output
    
    @staticmethod
    def backward(ctx, grad_output):
        # Извлекаем сохраненные тензоры и параметры
        input, weight, bias = ctx.saved_tensors
        stride, padding = ctx.stride, ctx.padding
        grad_input = grad_weight = grad_bias = None
        
        # Обратное масштабирование в обратном проходе
        channel_means = grad_output.mean(dim=(2, 3), keepdim=True)
        grad_output = grad_output / (channel_means + 1e-5)
        
        # Вычисляем градиенты для входного тензора, весов и смещения
        if ctx.needs_input_grad[0]:
            grad_input = torch.nn.grad.conv2d_input(input.shape, weight, grad_output, stride=stride, padding=padding)
        if ctx.needs_input_grad[1]:
            grad_weight = torch.nn.grad.conv2d_weight(input, weight.shape, grad_output, stride=stride, padding=padding)
        if ctx.needs_input_grad[2] and bias is not None:
            grad_bias = grad_output.sum(dim=(0, 2, 3))
        
        return grad_input, grad_weight, grad_bias, None, None


# Пользовательский сверточный слой, использующий CustomConvFunction
class CustomConv2d(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, stride=1, padding=1):
        super().__init__()
        # Инициализируем веса и смещения как обучаемые параметры
        self.weight = nn.Parameter(torch.randn(out_channels, in_channels, kernel_size, kernel_size))
        self.bias = nn.Parameter(torch.randn(out_channels))
        self.stride = stride
        self.padding = padding
    
    def forward(self, x):
        return CustomConvFunction.apply(x, self.weight, self.bias, self.stride, self.padding)


# Механизм внимания к каналам (Channel Attention)
class ChannelAttention(nn.Module):
    def __init__(self, in_channels, reduction=16):
        super().__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1) # Глобальный усредняющий пулинг
        self.fc1 = nn.Linear(in_channels, in_channels // reduction) # Первый полносвязный слой
        self.fc2 = nn.Linear(in_channels // reduction, in_channels) # Второй полносвязный слой
    
    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.avg_pool(x).view(b, c) # Сжимаем пространственные измерения
        y = F.relu(self.fc1(y)) # ReLU активация
        y = torch.sigmoid(self.fc2(y)).view(b, c, 1, 1) # Sigmoid активация и изменение формы
        return x * y # Масштабируем входные признаки весами внимания


# Пользовательская функция активации (масштабированный гиперболический тангенс)
class CustomActivationFunction(Function):
    @staticmethod
    def forward(ctx, input):
        ctx.save_for_backward(input) # Сохраняем вход для обратного прохода
        # Пользовательская активация: масштабированный tanh
        output = torch.tanh(input) * 1.5
        return output
    
    @staticmethod
    def backward(ctx, grad_output):
        input, = ctx.saved_tensors
        # Вычисляем градиент для масштабированного tanh
        grad_input = grad_output * 1.5 * (1 - torch.tanh(input) ** 2)
        return grad_input


# Пользовательский слой активации
class CustomActivation(nn.Module):
    def forward(self, x):
        return CustomActivationFunction.apply(x)


import torch
import torch.nn.functional as F
from torch.autograd import Function

class CustomPoolFunction(Function):
    @staticmethod
    def forward(ctx, input, kernel_size=2, stride=2):
        """
        Прямой проход: применить максимальное объединение и добавить масштабированный средний член.
        """
        ctx.kernel_size = kernel_size
        ctx.stride = stride
        output, indices = F.max_pool2d(input, kernel_size, stride, return_indices=True)
        mean_term = 0.1 * input.mean(dim=(2, 3), keepdim=True)
        output = output + mean_term
        ctx.save_for_backward(input, indices, mean_term)
        return output
    
    @staticmethod
    def backward(ctx, grad_output):
        """
        Обратный проход: вычисление градиентов для входных данных.
        """
        input, indices, mean_term = ctx.saved_tensors
        grad_input_maxpool = F.max_unpool2d(
            grad_output, 
            indices, 
            kernel_size=ctx.kernel_size, 
            stride=ctx.stride, 
            output_size=input.size()
        )
        grad_input_mean = 0.1 * grad_output.mean(dim=(2, 3), keepdim=True).expand_as(input)
        grad_input = grad_input_maxpool + grad_input_mean
        return grad_input, None, None


class CustomPool2d(nn.Module):
    def __init__(self, kernel_size=2, stride=2):
        super().__init__()
        self.kernel_size = kernel_size
        self.stride = stride
    
    def forward(self, x):
        return CustomPoolFunction.apply(x, self.kernel_size, self.stride)

# Пользовательская CNN с пользовательскими слоями
class CustomCNN(nn.Module):
    def __init__(self, input_channels=3, num_classes=10):
        super().__init__()
        self.conv1 = CustomConv2d(input_channels, 32, 3) # Пользовательский сверточный слой
        self.attention = ChannelAttention(32) # Слой внимания к каналам
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1) # Стандартный сверточный слой
        self.pool = CustomPool2d(2, 2) # Пользовательский слой пулинга
        self.activation = CustomActivation() # Пользовательский слой активации
        self.bn1 = nn.BatchNorm2d(32) # Пакетная нормализация
        self.bn2 = nn.BatchNorm2d(64) # Пакетная нормализация
        # Исправлено: 32x32 -> pool (16x16) -> pool (8x8). Размер входного тензора для FC должен быть 64 * 8 * 8
        self.fc = nn.Linear(64 * 8 * 8, num_classes) # Полносвязный слой
        self.dropout = nn.Dropout(0.25) # Слой Dropout
    
    def forward(self, x):
        x = self.activation(self.bn1(self.conv1(x))) # CustomConv -> BatchNorm -> CustomActivation
        x = self.attention(x) # Применяем внимание к каналам
        x = self.pool(x) # CustomPool
        x = self.activation(self.bn2(self.conv2(x))) # Conv -> BatchNorm -> CustomActivation
        x = self.pool(x) # CustomPool
        x = x.view(x.size(0), -1) # Выпрямляем тензор
        x = self.dropout(x) # Dropout
        x = self.fc(x) # Полносвязный слой
        return x


# 3.2 Остаточные блоки
# Базовый остаточный блок (Basic Residual Block)
class BasicResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, 3, stride, 1, bias=False) # Первая свертка
        self.bn1 = nn.BatchNorm2d(out_channels) # Пакетная нормализация
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, 1, 1, bias=False) # Вторая свертка
        self.bn2 = nn.BatchNorm2d(out_channels) # Пакетная нормализация
        self.shortcut = nn.Sequential() # Ярлык (ветка пропуска)
        # Если размеры не совпадают, используем свертку 1x1 для ярлыка
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 1, stride, bias=False),
                nn.BatchNorm2d(out_channels)
            )
    
    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x))) # Свертка -> BatchNorm -> ReLU
        out = self.bn2(self.conv2(out)) # Свертка -> BatchNorm
        out += self.shortcut(x) # Добавляем ярлык
        out = F.relu(out) # ReLU
        return out


# Остаточный блок "бутылочное горлышко" (Bottleneck Residual Block)
class BottleneckResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1):
        super().__init__()
        bottleneck_channels = out_channels // 4 # Уменьшение количества каналов
        self.conv1 = nn.Conv2d(in_channels, bottleneck_channels, 1, bias=False) # Свертка 1x1
        self.bn1 = nn.BatchNorm2d(bottleneck_channels)
        self.conv2 = nn.Conv2d(bottleneck_channels, bottleneck_channels, 3, stride, 1, bias=False) # Свертка 3x3
        self.bn2 = nn.BatchNorm2d(bottleneck_channels)
        self.conv3 = nn.Conv2d(bottleneck_channels, out_channels, 1, bias=False) # Свертка 1x1 для восстановления каналов
        self.bn3 = nn.BatchNorm2d(out_channels)
        self.shortcut = nn.Sequential()
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 1, stride, bias=False),
                nn.BatchNorm2d(out_channels)
            )
    
    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = F.relu(self.bn2(self.conv2(out)))
        out = self.bn3(self.conv3(out))
        out += self.shortcut(x)
        out = F.relu(out)
        return out


# Широкий остаточный блок (Wide Residual Block)
class WideResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1, width_factor=2):
        super().__init__()
        wide_channels = out_channels * width_factor # Увеличение количества каналов
        self.conv1 = nn.Conv2d(in_channels, wide_channels, 3, stride, 1, bias=False) # Первая свертка
        self.bn1 = nn.BatchNorm2d(wide_channels)
        self.conv2 = nn.Conv2d(wide_channels, out_channels, 3, 1, 1, bias=False) # Вторая свертка
        self.bn2 = nn.BatchNorm2d(out_channels)
        self.shortcut = nn.Sequential()
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 1, stride, bias=False),
                nn.BatchNorm2d(out_channels)
            )
    
    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += self.shortcut(x)
        out = F.relu(out)
        return out


# CNN с остаточными блоками
class ResidualCNN(nn.Module):
    def __init__(self, block_type, input_channels=3, num_classes=10):
        super().__init__()
        self.conv1 = nn.Conv2d(input_channels, 32, 3, 1, 1) # Первый сверточный слой
        self.bn1 = nn.BatchNorm2d(32) # Пакетная нормализация
        self.block1 = block_type(32, 32) # Первый остаточный блок
        self.block2 = block_type(32, 64, stride=2) # Второй остаточный блок с уменьшением размера
        self.pool = nn.AdaptiveAvgPool2d((4, 4)) # Адаптивный усредняющий пулинг
        self.fc = nn.Linear(64 * 4 * 4, num_classes) # Полносвязный слой
    
    def forward(self, x):
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.block1(x)
        x = self.block2(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1) # Выпрямляем тензор
        x = self.fc(x) # Полносвязный слой
        return x


# Вычисление нормы градиентов
def compute_gradient_norms(model):
    total_norm = 0.0
    for p in model.parameters():
        if p.grad is not None:
            param_norm = p.grad.data.norm(2) # L2-норма градиента
            total_norm += param_norm.item() ** 2
    return np.sqrt(total_norm) # Возвращаем общую норму


# Функция для тестирования пользовательских слоев
def test_custom_layers():
    print("Проверка CustomConv2d:")
    x = torch.randn(1, 3, 32, 32) # Случайный входной тензор
    custom_conv = CustomConv2d(3, 16, 3)
    standard_conv = nn.Conv2d(3, 16, 3, padding=1)
    custom_out = custom_conv(x)
    standard_out = standard_conv(x)
    print(f"Форма вывода CustomConv2d: {custom_out.shape}")
    print(f"Форма вывода Standard Conv2d: {standard_out.shape}")
    
    print("\nПроверка ChannelAttention:")
    attention = ChannelAttention(16)
    attn_out = attention(custom_out)
    print(f"Форма вывода ChannelAttention: {attn_out.shape}")
    
    print("\nПроверка CustomActivation:")
    activation = CustomActivation()
    act_out = activation(custom_out)
    print(f"Форма вывода CustomActivation: {act_out.shape}")
    
    print("\nПроверка CustomPool2d:")
    pool = CustomPool2d(2, 2)
    pool_out = pool(custom_out)
    print(f"Форма вывода CustomPool2d: {pool_out.shape}")


def main():
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Используемое устройство: {device}")
    
    # Тестирование пользовательских слоев
    print("\n=== Тестирование пользовательских слоев ===")
    test_custom_layers()
    
    # Загрузка набора данных CIFAR-10
    train_loader, test_loader = get_cifar_loaders(batch_size=64)
    
    # 3.1 Эксперимент с пользовательскими слоями
    print("\n=== Эксперимент с пользовательскими слоями ===")
    # Создаем экземпляр пользовательской CNN
    custom_cnn = CustomCNN(input_channels=3, num_classes=10).to(device)
    # Создаем экземпляр стандартной CNN (используем BasicResidualBlock как базовую для сравнения)
    standard_cnn = ResidualCNN(BasicResidualBlock, input_channels=3, num_classes=10).to(device)
    
    models = {'CustomCNN': custom_cnn, 'StandardCNN': standard_cnn}
    histories = {}
    training_times = {}
    
    for name, model in models.items():
        print(f"\nОбучение {name}...")
        start_time = time.time()
        # Обучаем модель и получаем историю
        history = train_model(model, train_loader, test_loader, epochs=5, lr=0.001, device=str(device))
        training_time = time.time() - start_time
        
        histories[name] = history
        training_times[name] = training_time
        
        print(f"Параметры: {count_parameters(model)}")
        print(f"Время обучения: {training_time:.2f} секунд")
        plot_training_history(history) # Строим график истории обучения
    
    # Сравнение пользовательской и стандартной CNN
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(histories['CustomCNN']['test_accs'], label='Custom CNN', marker='o')
    plt.plot(histories['StandardCNN']['test_accs'], label='Standard CNN', marker='s')
    plt.title('Сравнение точности на тестовой выборке')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    plt.plot(histories['CustomCNN']['test_losses'], label='Custom CNN', marker='o')
    plt.plot(histories['StandardCNN']['test_losses'], label='Standard CNN', marker='s')
    plt.title('Сравнение потерь на тестовой выборке')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # 3.2 Эксперимент с остаточными блоками
    print("\n=== Эксперимент с остаточными блоками ===")
    block_types = [
        (BasicResidualBlock, "Базовый остаточный"),
        (BottleneckResidualBlock, "Остаточный (бутылочное горлышко)"),
        (WideResidualBlock, "Широкий остаточный")
    ]
    
    histories_res = {}
    training_times_res = {}
    gradient_norms_res = {}
    
    for block_type, name in block_types:
        print(f"\nОбучение {name}...")
        model = ResidualCNN(block_type, input_channels=3, num_classes=10).to(device)
        
        # Отслеживание градиентов
        gradient_norms = []
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)
        model.train()
        
        for epoch in range(5): # Выполняем 5 эпох для сбора норм градиентов
            epoch_grad_norms = []
            for data, target in train_loader:
                data, target = data.to(device), target.to(device)
                optimizer.zero_grad()
                output = model(data)
                loss = criterion(output, target)
                loss.backward()
                epoch_grad_norms.append(compute_gradient_norms(model))
                optimizer.step()
            gradient_norms.append(np.mean(epoch_grad_norms))
        
        start_time = time.time()
        # Обучаем модель и получаем историю
        history = train_model(model, train_loader, test_loader, epochs=5, lr=0.001, device=str(device))
        training_time = time.time() - start_time
        
        histories_res[name] = history
        training_times_res[name] = training_time
        gradient_norms_res[name] = gradient_norms # Сохраняем нормы градиентов
        
        print(f"Параметры: {count_parameters(model)}")
        print(f"Время обучения: {training_time:.2f} секунд")
        plot_training_history(history)
        
        # Построение графика норм градиентов
        plt.figure(figsize=(8, 4))
        plt.plot(gradient_norms, label=f'Норма градиентов {name}')
        plt.title(f'Нормы градиентов во время обучения ({name})')
        plt.xlabel('Эпоха')
        plt.ylabel('Норма градиентов')
        plt.legend()
        plt.grid(True)
        plt.show()
    
    # Сравнение остаточных блоков
    for name1, name2 in [('Базовый остаточный', 'Остаточный (бутылочное горлышко)'), 
                         ('Остаточный (бутылочное горлышко)', 'Широкий остаточный')]:
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.plot(histories_res[name1]['test_accs'], label=name1, marker='o')
        plt.plot(histories_res[name2]['test_accs'], label=name2, marker='s')
        plt.title('Сравнение точности на тестовой выборке')
        plt.legend()
        plt.grid(True)
        
        plt.subplot(1, 2, 2)
        plt.plot(histories_res[name1]['test_losses'], label=name1, marker='o')
        plt.plot(histories_res[name2]['test_losses'], label=name2, marker='s')
        plt.title('Сравнение потерь на тестовой выборке')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()


if __name__ == "__main__":
    main()
```
