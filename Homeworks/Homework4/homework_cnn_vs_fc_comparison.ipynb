```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from datasets import get_mnist_loaders, get_cifar_loaders
from models import SimpleCNN, CNNWithResidual, ResidualBlock
from trainer import train_model
from utils import plot_training_history, count_parameters, compare_models 
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
from torch.nn.utils import clip_grad_norm_


# Полносвязная нейронная сеть
class FullyConnectedNet(nn.Module):
    def __init__(self, input_size=28*28, num_classes=10):
        super().__init__()
        self.flatten = nn.Flatten() # Слой для выпрямления входных данных
        self.fc1 = nn.Linear(input_size, 512) # Первый полносвязный слой
        self.fc2 = nn.Linear(512, 256) # Второй полносвязный слой
        self.fc3 = nn.Linear(256, 128) # Третий полносвязный слой
        self.fc4 = nn.Linear(128, num_classes) # Выходной полносвязный слой
        self.dropout = nn.Dropout(0.25) # Слой Dropout для регуляризации
    
    def forward(self, x):
        x = self.flatten(x) # Выпрямляем вход
        x = torch.relu(self.fc1(x)) # Применяем ReLU к первому слою
        x = self.dropout(x) # Применяем Dropout
        x = torch.relu(self.fc2(x)) # Применяем ReLU ко второму слою
        x = self.dropout(x) # Применяем Dropout
        x = torch.relu(self.fc3(x)) # Применяем ReLU к третьему слою
        x = self.dropout(x) # Применяем Dropout
        x = self.fc4(x) # Пропускаем через выходной слой
        return x


# Сверточная нейронная сеть с остаточными соединениями и регуляризацией
class RegularizedCNNWithResidual(nn.Module):
    def __init__(self, input_channels=3, num_classes=10):
        super().__init__()
        self.conv1 = nn.Conv2d(input_channels, 32, 3, 1, 1) # Первый сверточный слой
        self.bn1 = nn.BatchNorm2d(32) # Пакетная нормализация
        self.res1 = ResidualBlock(32, 32) # Первый остаточный блок
        self.res2 = ResidualBlock(32, 64, 2) # Второй остаточный блок (с изменением размера)
        self.res3 = ResidualBlock(64, 64) # Третий остаточный блок
        self.pool = nn.AdaptiveAvgPool2d((4, 4)) # Адаптивный слой усредняющего пулинга
        self.fc = nn.Linear(64 * 4 * 4, num_classes) # Полносвязный слой
        self.dropout = nn.Dropout(0.3) # Слой Dropout
    
    def forward(self, x):
        x = torch.relu(self.bn1(self.conv1(x))) # Свертка, BatchNorm, ReLU
        x = self.res1(x) # Остаточный блок 1
        x = self.dropout(x) # Dropout
        x = self.res2(x) # Остаточный блок 2
        x = self.res3(x) # Остаточный блок 3
        x = self.pool(x) # Усредняющий пулинг
        x = x.view(x.size(0), -1) # Выпрямляем для полносвязного слоя
        x = self.dropout(x) # Dropout
        x = self.fc(x) # Выходной полносвязный слой
        return x


# Вычисление нормы градиентов
def compute_gradient_norms(model):
    total_norm = 0.0
    for p in model.parameters():
        if p.grad is not None:
            param_norm = p.grad.data.norm(2) # L2-норма градиента параметра
            total_norm += param_norm.item() ** 2
    return np.sqrt(total_norm) # Возвращаем общую норму градиентов


# Получение матрицы ошибок
def get_confusion_matrix(model, data_loader, device):
    model.eval() # Переводим модель в режим оценки
    all_preds = []
    all_targets = []
    with torch.no_grad(): # Отключаем вычисление градиентов
        for data, target in data_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            preds = output.argmax(dim=1) # Получаем предсказанные классы
            all_preds.extend(preds.cpu().numpy()) # Добавляем предсказания в список
            all_targets.extend(target.cpu().numpy()) # Добавляем истинные метки в список
    return confusion_matrix(all_targets, all_preds) # Возвращаем матрицу ошибок


# Построение матрицы ошибок
def plot_confusion_matrix(cm, title, classes):
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes) # Используем seaborn для красивой визуализации
    plt.title(title) # Заголовок графика
    plt.xlabel('Предсказанные классы') # Подпись оси X
    plt.ylabel('Истинные классы') # Подпись оси Y
    plt.show()


def main():
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Используемое устройство: {device}")

    # Эксперимент с MNIST
    print("\n=== Эксперимент с MNIST ===")
    train_loader_mnist, test_loader_mnist = get_mnist_loaders(batch_size=64)
    
    # Инициализация моделей
    fc_net_mnist = FullyConnectedNet(input_size=28*28, num_classes=10).to(device)
    simple_cnn_mnist = SimpleCNN(input_channels=1, num_classes=10).to(device)
    residual_cnn_mnist = CNNWithResidual(input_channels=1, num_classes=10).to(device)

    models_mnist = {
        'FullyConnected': fc_net_mnist,
        'SimpleCNN': simple_cnn_mnist,
        'ResidualCNN': residual_cnn_mnist
    }

    histories_mnist = {}
    training_times_mnist = {}
    inference_times_mnist = {}

    # Обучение и оценка
    for name, model in models_mnist.items():
        print(f"\nОбучение {name}...")
        start_time = time.time()
        # Предполагается, что train_model возвращает историю обучения
        history = train_model(model, train_loader_mnist, test_loader_mnist, epochs=5, lr=0.001, device=str(device))
        training_time = time.time() - start_time
        
        # Измерение времени инференса
        model.eval() # Переводим модель в режим оценки
        inference_start = time.time()
        with torch.no_grad():
            for data, _ in test_loader_mnist:
                data = data.to(device)
                model(data)
        inference_time = (time.time() - inference_start) / len(test_loader_mnist.dataset) # Время на один образец
        
        histories_mnist[name] = history
        training_times_mnist[name] = training_time
        inference_times_mnist[name] = inference_time
        
        print(f"Параметры: {count_parameters(model)}")
        print(f"Время обучения: {training_time:.2f} секунд")
        print(f"Время инференса на образец: {inference_time:.6f} секунд")
        
        plot_training_history(history) # Построение графика истории обучения
        
        # Построение матрицы ошибок
        cm = get_confusion_matrix(model, test_loader_mnist, device)
        plot_confusion_matrix(cm, f"Матрица ошибок {name} (MNIST)", range(10))

    # Сравнение моделей
    compare_models(histories_mnist['FullyConnected'], histories_mnist['SimpleCNN'])
    compare_models(histories_mnist['SimpleCNN'], histories_mnist['ResidualCNN'])

    # Эксперимент с CIFAR-10
    print("\n=== Эксперимент с CIFAR-10 ===")
    train_loader_cifar, test_loader_cifar = get_cifar_loaders(batch_size=64)
    
    # Инициализация моделей
    fc_net_cifar = FullyConnectedNet(input_size=32*32*3, num_classes=10).to(device) # Для CIFAR-10 размер входного изображения 32x32x3
    residual_cnn_cifar = CNNWithResidual(input_channels=3, num_classes=10).to(device)
    reg_residual_cnn_cifar = RegularizedCNNWithResidual(input_channels=3, num_classes=10).to(device)

    models_cifar = {
        'FullyConnected': fc_net_cifar,
        'ResidualCNN': residual_cnn_cifar,
        'RegularizedResidualCNN': reg_residual_cnn_cifar
    }

    histories_cifar = {}
    training_times_cifar = {}
    inference_times_cifar = {}
    gradient_norms_cifar = {}

    # Обучение и оценка с отслеживанием градиентов
    for name, model in models_cifar.items():
        print(f"\nОбучение {name}...")
        start_time = time.time()
        gradient_norms = []
        
        # Модифицированный цикл обучения для отслеживания градиентов
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)
        model.train()
        
        # Передаем train_loader, test_loader, epochs, device в train_model
        # Возможно, стоит пересмотреть, как интегрируется отслеживание градиентов,
        # если train_model уже сам выполняет полный цикл обучения.
        # Для этого примера будем считать, что этот цикл используется для отслеживания градиентов,
        # а затем вызывается train_model для получения полной истории.
        for epoch in range(5): # Отдельный цикл для сбора норм градиентов
            epoch_grad_norms = []
            for data, target in train_loader_cifar:
                data, target = data.to(device), target.to(device)
                optimizer.zero_grad()
                output = model(data)
                loss = criterion(output, target)
                loss.backward()
                epoch_grad_norms.append(compute_gradient_norms(model)) # Вычисляем норму градиентов
                clip_grad_norm_(model.parameters(), max_norm=1.0) # Отсечение градиентов
                optimizer.step()
            gradient_norms.append(np.mean(epoch_grad_norms)) # Средняя норма градиентов за эпоху
        
        # Теперь вызываем train_model для получения полной истории обучения,
        # с учетом того, что модель уже немного обучена в предыдущем цикле.
        # В реальном сценарии, отслеживание градиентов должно быть частью train_model.
        history = train_model(model, train_loader_cifar, test_loader_cifar, epochs=5, lr=0.001, device=str(device))
        training_time = time.time() - start_time
        
        # Измерение времени инференса
        model.eval()
        inference_start = time.time()
        with torch.no_grad():
            for data, _ in test_loader_cifar:
                data = data.to(device)
                model(data)
        inference_time = (time.time() - inference_start) / len(test_loader_cifar.dataset)
        
        histories_cifar[name] = history
        training_times_cifar[name] = training_time
        inference_times_cifar[name] = inference_time
        gradient_norms_cifar[name] = gradient_norms # Сохраняем нормы градиентов
        
        print(f"Параметры: {count_parameters(model)}")
        print(f"Время обучения: {training_time:.2f} секунд")
        print(f"Время инференса на образец: {inference_time:.6f} секунд")
        
        plot_training_history(history)
        
        # Построение матрицы ошибок
        cm = get_confusion_matrix(model, test_loader_cifar, device)
        cifar_classes = ['самолет', 'автомобиль', 'птица', 'кошка', 'олень', 'собака', 'лягушка', 'лошадь', 'корабль', 'грузовик']
        plot_confusion_matrix(cm, f"Матрица ошибок {name} (CIFAR-10)", cifar_classes)
        
        # Построение графика норм градиентов
        plt.figure(figsize=(8, 4))
        plt.plot(gradient_norms, label=f'Норма градиентов {name}')
        plt.title(f'Нормы градиентов во время обучения ({name})')
        plt.xlabel('Эпоха')
        plt.ylabel('Норма градиентов')
        plt.legend()
        plt.grid(True)
        plt.show()

    # Сравнение моделей
    compare_models(histories_cifar['FullyConnected'], histories_cifar['ResidualCNN'])
    compare_models(histories_cifar['ResidualCNN'], histories_cifar['RegularizedResidualCNN'])

if __name__ == "__main__":
    main()
```
