```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
import time
from datasets import get_cifar_loaders
from models import ResidualBlock
from trainer import train_model
from utils import plot_training_history, count_parameters


# Определение класса CNNKernelSize для анализа влияния размера ядра свертки
class CNNKernelSize(nn.Module):
    def __init__(self, kernel_config, input_channels=3, num_classes=10):
        super().__init__()
        self.layers = nn.ModuleList()
        in_channels = input_channels
        out_channels = [32, 64, 128]  # Фиксированное увеличение количества каналов
        
        for i, kernel_size in enumerate(kernel_config):
            if isinstance(kernel_size, tuple):  # Для комбинации 1x1 + 3x3
                # Слой 1x1 для уменьшения количества каналов
                self.layers.append(nn.Conv2d(in_channels, out_channels[i]//2, 1))
                self.layers.append(nn.BatchNorm2d(out_channels[i]//2))
                self.layers.append(nn.ReLU())
                # Слой 3x3 для основной свертки
                self.layers.append(nn.Conv2d(out_channels[i]//2, out_channels[i], 3, padding=1))
            else:
                # Обычный сверточный слой с заданным размером ядра
                self.layers.append(nn.Conv2d(in_channels, out_channels[i], kernel_size, padding=kernel_size//2))
            self.layers.append(nn.BatchNorm2d(out_channels[i])) # Пакетная нормализация
            self.layers.append(nn.ReLU()) # Функция активации ReLU
            self.layers.append(nn.MaxPool2d(2, 2)) # Макс-пулинг для уменьшения размера
            in_channels = out_channels[i] # Обновляем количество входных каналов для следующего слоя
        
        self.pool = nn.AdaptiveAvgPool2d((4, 4)) # Адаптивный слой усредняющего пулинга
        self.fc = nn.Linear(128 * 4 * 4, num_classes) # Полносвязный слой для классификации
        self.dropout = nn.Dropout(0.25) # Слой Dropout для регуляризации
    
    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        x = self.pool(x) # Применяем усредняющий пулинг
        x = x.view(x.size(0), -1) # Выпрямляем тензор для полносвязного слоя
        x = self.dropout(x) # Применяем Dropout
        x = self.fc(x) # Пропускаем через полносвязный слой
        return x
    
    def get_first_layer_activations(self, x):
        # Получаем активации после первого сверточного блока
        x = self.layers[0](x)  # Первая свертка
        x = self.layers[1](x)  # Пакетная нормализация
        x = self.layers[2](x)  # ReLU
        return x


# Определение класса CNNDepth для анализа влияния глубины сети
class CNNDepth(nn.Module):
    def __init__(self, num_conv_layers, use_residual=False, input_channels=3, num_classes=10):
        super().__init__()
        self.use_residual = use_residual
        self.layers = nn.ModuleList()
        in_channels = input_channels
        out_channels = [32, 64, 128, 256, 256, 256][:num_conv_layers]  # Убрано max(num_conv_layers, 3)
        
        for i in range(num_conv_layers):
            if use_residual and i > 0 and i % 2 == 0:
                self.layers.append(ResidualBlock(in_channels, out_channels[i]))
                in_channels = out_channels[i]
            else:
                self.layers.append(nn.Conv2d(in_channels, out_channels[i], 3, padding=1))
                self.layers.append(nn.BatchNorm2d(out_channels[i]))
                self.layers.append(nn.ReLU())
                if i % 2 == 1:
                    self.layers.append(nn.MaxPool2d(2, 2))
                in_channels = out_channels[i]
        
        self.pool = nn.AdaptiveAvgPool2d((4, 4))
        # Рассчитываем размер входа в полносвязный слой
        final_channels = out_channels[-1]
        self.fc = nn.Linear(final_channels * 4 * 4, num_classes)
        self.dropout = nn.Dropout(0.25)
    
    def forward(self, x):
        for i, layer in enumerate(self.layers):
            x = layer(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = self.dropout(x)
        x = self.fc(x)
        return x
    
    def get_feature_maps(self, x, layer_idx):
        for i, layer in enumerate(self.layers):
            x = layer(x)
            if i == layer_idx:
                return x
        return x


# Вычисление нормы градиентов
def compute_gradient_norms(model):
    total_norm = 0.0
    for p in model.parameters():
        if p.grad is not None:
            param_norm = p.grad.data.norm(2) # L2-норма градиента параметра
            total_norm += param_norm.item() ** 2
    return np.sqrt(total_norm) # Возвращаем общую норму градиентов


# Вычисление рецептивного поля
def compute_receptive_field(kernel_sizes, strides):
    rf = 1
    # Формула для вычисления рецептивного поля
    for k, s in zip(kernel_sizes, strides):
        rf += (k - 1) * s
    return rf


# Визуализация активаций/карт признаков
def visualize_activations(activations, title, num_filters=16):
    # Преобразуем тензор активаций в массив numpy и выбираем первый образец
    activations = activations.detach().cpu().numpy()[0]
    fig, axes = plt.subplots(4, 4, figsize=(12, 12))
    for i, ax in enumerate(axes.flat):
        if i < num_filters: # Отображаем до num_filters карт
            ax.imshow(activations[i], cmap='viridis') # Используем цветовую карту viridis
            ax.axis('off') # Отключаем оси
    plt.suptitle(title) # Общий заголовок для графиков
    plt.tight_layout() # Автоматическая настройка макета
    plt.show()


def main():
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Используемое устройство: {device}")
    
    train_loader, test_loader = get_cifar_loaders(batch_size=64)
    
    # ---
    ## 2.1 Анализ размера ядра свертки
    # ---
    print("\n=== Анализ размера ядра свертки ===")
    kernel_configs = [
        ([3, 3, 3], "Ядра 3x3"),
        ([5, 5, 5], "Ядра 5x5"),
        ([7, 7, 7], "Ядра 7x7"),
        ([(1, 3), 3, 3], "Комбинация 1x1 + 3x3") # Пример использования кортежа для 1x1 + 3x3
    ]
    
    histories_kernel = {}
    training_times_kernel = {}
    inference_times_kernel = {}
    receptive_fields = []
    
    for config, name in kernel_configs:
        print(f"\nОбучение CNN с {name}...")
        model = CNNKernelSize(config, input_channels=3, num_classes=10).to(device)
        
        start_time = time.time()
        # Обучаем модель и получаем историю
        history = train_model(model, train_loader, test_loader, epochs=5, lr=0.001, device=str(device))
        training_time = time.time() - start_time
        
        model.eval() # Переводим модель в режим оценки
        inference_start = time.time()
        with torch.no_grad():
            for data, _ in test_loader:
                data = data.to(device)
                model(data)
        inference_time = (time.time() - inference_start) / len(test_loader.dataset) # Время инференса на один образец
        
        histories_kernel[name] = history
        training_times_kernel[name] = training_time
        inference_times_kernel[name] = inference_time
        
        print(f"Параметры: {count_parameters(model)}")
        print(f"Время обучения: {training_time:.2f} секунд")
        print(f"Время инференса на образец: {inference_time:.6f} секунд")
        
        # Вычисление и вывод рецептивного поля
        # Для 1x1 + 3x3 комбинации используем 3 для вычисления RF
        kernel_sizes_for_rf = [k if isinstance(k, int) else 3 for k in config]
        rf = compute_receptive_field(kernel_sizes_for_rf, [1, 1, 1]) # Предполагаем шаг 1 для каждого слоя пулинга
        receptive_fields.append((name, rf))
        print(f"Рецептивное поле: {rf}")
        
        # Визуализация активаций первого слоя
        with torch.no_grad():
            for data, _ in test_loader:
                data = data[:1].to(device) # Берем один образец
                activations = model.get_first_layer_activations(data)
                visualize_activations(activations, f"Активации первого слоя ({name})")
                break # Выходим после первого пакета
        
        plot_training_history(history) # Построение графика истории обучения
    
    # Сравнение результатов для разных размеров ядер
    for name1, name2 in [('Ядра 3x3', 'Ядра 5x5'), ('Ядра 5x5', 'Ядра 7x7'), ('Ядра 3x3', 'Комбинация 1x1 + 3x3')]:
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.plot(histories_kernel[name1]['test_accs'], label=name1, marker='o')
        plt.plot(histories_kernel[name2]['test_accs'], label=name2, marker='s')
        plt.title('Сравнение точности на тестовой выборке')
        plt.legend()
        plt.grid(True)
        
        plt.subplot(1, 2, 2)
        plt.plot(histories_kernel[name1]['test_losses'], label=name1, marker='o')
        plt.plot(histories_kernel[name2]['test_losses'], label=name2, marker='s')
        plt.title('Сравнение потерь на тестовой выборке')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    
    # ---
    ## 2.2 Анализ глубины сети
    # ---
    print("\n=== Анализ глубины сети ===")
    depth_configs = [
        (2, False, "Мелкая CNN (2 сверточных слоя)"),
        (4, False, "Средняя CNN (4 сверточных слоя)"),
        (6, False, "Глубокая CNN (6 сверточных слоев)"),
        (6, True, "Остаточная CNN (6 сверточных слоев)") # С остаточными соединениями
    ]
    
    histories_depth = {}
    training_times_depth = {}
    inference_times_depth = {}
    gradient_norms_depth = {}
    
    for num_layers, use_residual, name in depth_configs:
        print(f"\nОбучение {name}...")
        model = CNNDepth(num_layers, use_residual, input_channels=3, num_classes=10).to(device)
        
        # Отслеживание градиентов
        gradient_norms = []
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)
        model.train()
        
        for epoch in range(5): # Выполняем 5 эпох для сбора норм градиентов
            epoch_grad_norms = []
            for data, target in train_loader:
                data, target = data.to(device), target.to(device)
                optimizer.zero_grad()
                output = model(data)
                loss = criterion(output, target)
                loss.backward()
                epoch_grad_norms.append(compute_gradient_norms(model)) # Вычисляем норму градиентов
                optimizer.step()
            gradient_norms.append(np.mean(epoch_grad_norms)) # Средняя норма градиентов за эпоху
        
        start_time = time.time()
        # Обучаем модель и получаем историю. Важно: если train_model уже запускает 5 эпох,
        # то предыдущий цикл может быть избыточным, или train_model должен быть изменен
        # для сбора норм градиентов внутри себя. Здесь мы предполагаем, что train_model
        # продолжает обучение с того места, где остановился предыдущий цикл.
        history = train_model(model, train_loader, test_loader, epochs=5, lr=0.001, device=str(device))
        training_time = time.time() - start_time
        
        model.eval()
        inference_start = time.time()
        with torch.no_grad():
            for data, _ in test_loader:
                data = data.to(device)
                model(data)
        inference_time = (time.time() - inference_start) / len(test_loader.dataset)
        
        histories_depth[name] = history
        training_times_depth[name] = training_time
        inference_times_depth[name] = inference_time
        gradient_norms_depth[name] = gradient_norms # Сохраняем нормы градиентов
        
        print(f"Параметры: {count_parameters(model)}")
        print(f"Время обучения: {training_time:.2f} секунд")
        print(f"Время инференса на образец: {inference_time:.6f} секунд")
        
        # Визуализация карт признаков последнего сверточного слоя
        with torch.no_grad():
            for data, _ in test_loader:
                data = data[:1].to(device)
                # Выбираем активации предпоследнего слоя, чтобы увидеть выход последнего Conv/ResidualBlock перед пулингом
                # len(model.layers)-4 - это пример, может потребоваться корректировка в зависимости от точной структуры слоя
                feature_maps = model.get_feature_maps(data, len(model.layers)-4) 
                visualize_activations(feature_maps, f"Карты признаков последнего слоя ({name})")
                break
        
        plot_training_history(history)
        
        # Построение графика норм градиентов
        plt.figure(figsize=(8, 4))
        plt.plot(gradient_norms, label=f'Норма градиентов {name}')
        plt.title(f'Нормы градиентов во время обучения ({name})')
        plt.xlabel('Эпоха')
        plt.ylabel('Норма градиентов')
        plt.legend()
        plt.grid(True)
        plt.show()
    
    # Сравнение результатов для разной глубины
    for name1, name2 in [('Мелкая CNN (2 сверточных слоя)', 'Средняя CNN (4 сверточных слоя)'), 
                         ('Средняя CNN (4 сверточных слоя)', 'Глубокая CNN (6 сверточных слоев)'), 
                         ('Глубокая CNN (6 сверточных слоев)', 'Остаточная CNN (6 сверточных слоев)')]:
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.plot(histories_depth[name1]['test_accs'], label=name1, marker='o')
        plt.plot(histories_depth[name2]['test_accs'], label=name2, marker='s')
        plt.title('Сравнение точности на тестовой выборке')
        plt.legend()
        plt.grid(True)
        
        plt.subplot(1, 2, 2)
        plt.plot(histories_depth[name1]['test_losses'], label=name1, marker='o')
        plt.plot(histories_depth[name2]['test_losses'], label=name2, marker='s')
        plt.title('Сравнение потерь на тестовой выборке')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()


if __name__ == "__main__":
    main()
```
